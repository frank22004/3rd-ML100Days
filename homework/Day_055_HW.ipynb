# 作業

- 作業目標 : iris (dataset.load_iris()) 資料嘗試跑 kmeans (可以測試不同的群數 , init 等)

```python
from sklearn import datasets
from sklearn.cluster import KMeans
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

iris = datasets.load_iris()
X = iris.data
y = iris.target
```

### 載入 相關套件 並 執行 kmean 實驗 ...

- 實驗測試不同的群數
- 實驗測試不同的初始值
- 呈現結果

### K = 8

```python
estimators = KMeans(n_clusters=8,n_init=10,init='random')
estimators.fit(X)
estimators.inertia_
```

```python
estimators.cluster_centers_
```

```python
estimators.labels_
```

```python
plt.scatter(X[:,0],X[:,1],c=estimators.labels_)
plt.scatter(estimators.cluster_centers_[:,0],estimators.cluster_centers_[:,1], c='red', s=100, alpha=0.5)
```

### k=8 with low initial

```python
estimators = KMeans(n_clusters=8,n_init=3,init='random')
estimators.fit(X)
estimators.inertia_
```

```python
plt.scatter(X[:,0],X[:,1],c=estimators.labels_)
plt.scatter(estimators.cluster_centers_[:,0],estimators.cluster_centers_[:,1], c='red', s=100, alpha=0.5)
```

High and low initial value shared the similar classification.
